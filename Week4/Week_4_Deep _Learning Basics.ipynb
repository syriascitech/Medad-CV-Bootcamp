## 1. كيف نُعلم الآلة؟ (البرمجة التقليدية vs التعلم العميق)

### أ- البرمجة التقليدية (Traditional Programming)
في البرمجة العادية، نقوم نحن كبشر بكتابة **القواعد (Rules)** يدوياً. الآلة تأخذ البيانات، تطبق عليها قواعدنا، وتعطينا النتيجة.

*   **مثال:** نظام حساب معدل الطالب.
    *   **المدخلات:** درجات المواد.
    *   **القاعدة اليدوية:** (مجموع الدرجات / عددها).
    *   **المخرج:** المعدل النهائي.


### ب- متى يفشل هذا النظام؟ (The Limitation)
تخيل لو أردنا كتابة قواعد لتمييز الرقم **(4)** المكتوب بخط اليد. هل سنكتب قاعدة تقول "إذا وجد خط مائل بزاوية 45 درجة..."؟ ماذا لو كان الخط منحنياً؟ أو كان الرقم مائلاً؟
هنا تصبح القواعد اليدوية مستحيلة التطبيق لكثرة الاحتمالات.

### ج- نموذج التعلم العميق (Deep Learning Paradigm)
في التعلم العميق، نقوم بعكس العملية. نحن نعطي الآلة **البيانات (Data)** و **النتائج (Outputs)**، وهي من يقوم باستنتاج **القواعد (Rules)** أو الأنماط.


### ما هو التعلم العميق (بالمستوى العالي)؟
هو فرع من الذكاء الاصطناعي يهدف لبناء "دماغ اصطناعي" (شبكة عصبية) قادر على تعلم تمثيل البيانات من خلال مستويات متعددة من التجريد. هو لا يحتاج منا إخباره أين الحافة أو أين الدائرة، بل يكتشف هذه الميزات تلقائياً من خلال التجربة والخطأ وتعديل نفسه ذاتياً ليصبح خبيراً في المهمة المطلوبة.

## المفاهيم الأساسية (Core Concepts)

| المفهوم | التعريف |
|---|---|
| **العصبون (Neuron)** | وحدة الحساب الأساسية التي تجمع المدخلات وتخرج إشارة واحدة. |
| **الوزن (Weight)** | يحدد مدى قوة أو أهمية مدخل معين بالنسبة للعصبون (ذاكرة الشبكة). |
| **الانحياز (Bias)** | قيمة مضافة تسمح بإزاحة دالة التنشيط لليسار أو اليمين لتحسين دقة التوقعات. |
| **طبقة الدخل (Input Layer)** | الطبقة الأولى التي تستقبل البيانات الخام (مثل بكسلات الصورة). |
| **الطبقات المخفية (Hidden Layers)** | الطبقات الوسطى حيث يتم استخلاص الأنماط والميزات المعقدة. |
| **طبقة الخرج (Output Layer)** | الطبقة النهائية التي تعطي النتيجة (تصنيف الرقم من 0 إلى 9). |
| **دالة التنشيط (Activation)** | دالة (مثل ReLU) تحول الجمع الخطي إلى مخرجات غير خطية لفهم الأنماط الصعبة. |
| **التمرير الأمامي (Forward Pass)** | رحلة البيانات من المدخلات عبر الطبقات وصولاً للنتيجة النهائية. |
| **الانتشار العكسي (Backpropagation)** | عملية حساب الخطأ والعودة للخلف لتعديل الأوزان (عملية التعلم). |

## ما هي الشبكة العصبية؟ (What is a Neural Network?)

الشبكة العصبية هي نموذج رياضي مستوحى من طريقة عمل الدماغ البشري. تتكون من وحدات معالجة صغيرة تسمى **العصبونات (Neurons)** مرتبطة ببعضها البعض لتمثيل علاقات معقدة في البيانات.

### 1. العصبون والأوزان (The Neuron & Weights)

العصبون هو الوحدة الأساسية؛ يستقبل مدخلات ($x$)، وكل مدخل له **وزن (Weight - $w$)** يحدد مدى أهميته.

![Neuron Structure](https://cdn.prod.website-files.com/680a070c3b99253410dd3df5/684d8626f3a3145136e891cd_681cc829292a5d61a39e0bd6_67ed55f61778c15890e01904_671b844911a414d4522be1b6_671b83aa76e220df9985dd00_Networks_fig1.png)

- **الأوزان ($w$):** هي "ذاكرة" الشبكة. أثناء التعلم، تقوم الشبكة بتعديل هذه الأوزان لتقليل الخطأ.
- **الانحياز ($b$):** قيمة إضافية تسمح للشبكة بضبط المخرجات حتى لو كانت المدخلات صفراً.

### 2. هيكلية الطبقات (Network Layers)

تتنظم العصبونات في ثلاثة أنواع من الطبقات:
1. **طبقة الدخل (Input Layer):** تستقبل البيانات الخام (مثل قيم بكسلات الصورة).
2. **الطبقات المخفية (Hidden Layers):** هنا يحدث السحر؛ حيث تستخرج الشبكة الميزات والأنماط المعقدة.
3. **طبقة الخرج (Output Layer):** تعطي النتيجة النهائية (مثل تصنيف الصورة كـ "رقم 5").

![Neural Network Layers](https://substackcdn.com/image/fetch/$s_!kwO_!,f_auto,q_auto:good,fl_progressive:steep/https%3A%2F%2Fsubstack-post-media.s3.amazonaws.com%2Fpublic%2Fimages%2Fc31161d1-4b31-4a04-8f28-b3e482df8a79_1280x720.gif)

---

### 3. التمرير الأمامي (Forward Propagation)

هي عملية تدفق البيانات من المدخلات عبر الطبقات المخفية وصولاً إلى المخرج. في كل عصبون:
1. يتم ضرب المدخلات بالأوزان وجمعها: $Z = (w_1x_1 + w_2x_2 + ... + b)$
2. يتم تمرير النتيجة عبر **دالة تنشيط** (مثل ReLU) لإضافة الخاصية غير الخطية.

### 4. التمرير الخلفي والتعلم (Backpropagation & Learning)

هذا هو الجزء المسؤول عن "الذكاء". بعد التمرير الأمامي، نقارن النتيجة بالجواب الصحيح باستخدام **دالة الخسارة (Loss Function)**. إذا كان هناك خطأ:
1. يتم حساب "مسؤولية" كل وزن عن هذا الخطأ (التفاضل).
2. نعود من المخرج إلى المدخلات (للخلف) لنعدل الأوزان قليلاً باستخدام **المحسن (Optimizer)** مثل Adam.

### 4. دور الأوزان (Weights) وآلية التنشيط

- **التنشيط:** عندما يتطابق الدخل البصري مع النموذج المخزن في الأوزان، يرتفع خرج العصبون (Activation)، معلناً اكتشاف الميزة.

### 5. طبقة الخرج (Output Layer): اتخاذ القرار
تمثل المرحلة النهائية حيث يوجد 10 عصبونات (من 0 إلى 9).
- **الاستدلال:** يقوم عصبون الرقم **9** بتقييم قوة الإشارات القادمة؛ فإذا وجد إشارة قوية لـ "دائرة" وإشارة قوية لـ "خط رأسي"، سيحقق أعلى قيمة تنشيط، ويكون هو القرار النهائي للنظام.

```python
# مثال برمجى بسيط يوضح عملية الجمع المرجح (Weighted Sum)
import numpy as np

# مدخلات تمثل بكسلات بسيطة
inputs = np.array([0.5, 0.8])
# أوزان افتراضية تعلمتها الشبكة
weights = np.array([0.4, -0.2])
# انحياز (Bias)
bias = 0.1

# التمرير الأمامي داخل عصبون واحد
weighted_sum = np.dot(inputs, weights) + bias
# دالة التنشيط (ReLU)
output = max(0, weighted_sum)

print(f'الجمع المرجح: {weighted_sum}')
print(f'خرج العصبون بعد ReLU: {output}')
```

### 1. مفهوم الطبقة (Layer)
في الواقع، لا يوجد عصبون وحيد، بل مجموعة عصبونات تعمل معاً. هنا نمثل طبقة تحتوي على 3 عصبونات تعالج نفس المدخلات.
</div>

```python
import numpy as np

# مدخلات الصورة (مثلاً قيمتين من البكسلات)
inputs = np.array([0.5, 0.8])

# مصفوفة الأوزان: كل عمود يمثل أوزان عصبون مختلف (لدينا 3 عصبونات هنا)
weights_layer = np.array([
    [0.2, 0.5, -0.1], # أوزان المدخل الأول لكل عصبون
    [0.7, -0.3, 0.4]  # أوزان المدخل الثاني لكل عصبون
])

# الانحياز لكل عصبون
biases = np.array([0.1, 0.2, -0.1])

# حساب خرج الطبقة كاملة بعملية ضرب مصفوفات واحدة
layer_outputs = np.dot(inputs, weights_layer) + biases
print(f'مخرجات الطبقة (3 عصبونات): {layer_outputs}')
```

<div dir="rtl">


### 2. الشبكة متعددة الطبقات (Deep Network Forward Pass)
هنا نرى كيف ينتقل المخرج من الطبقة الأولى ليصبح مدخلاً للطبقة الثانية (المخفية).

```python
# لنعتبر أن مخرجات الطبقة السابقة هي مدخلات للطبقة التالية
hidden_layer_input = layer_outputs

# أوزان الطبقة التالية (نحول الـ 3 مدخلات إلى مخرج واحد فقط)
weights_output_node = np.array([0.4, 0.9, -0.5])
bias_output = 0.5

# الحساب النهائي
final_result = np.dot(hidden_layer_input, weights_output_node) + bias_output
print(f'النتيجة النهائية بعد طبقتين: {final_result}')
```

### 3. تأثير تغيير الأوزان (Weights Sensitivity)
هذا الكود يوضح كيف أن تغيير الأوزان يغير قرار الشبكة تماماً، وهو جوهر عملية التعلم.

```python
def simple_predict(x, w):
    return "تنشيط عالي" if np.dot(x, w) > 0.5 else "تنشيط منخفض"

image_pixels = np.array([0.9, 0.1]) # بكسل ساطع وبكسل باهت

w_pattern1 = np.array([1.0, 0.0]) # وزن يهتم فقط بالبكسل الأول
w_pattern2 = np.array([0.0, 1.0]) # وزن يهتم فقط بالبكسل الثاني

print(f'النمط 1: {simple_predict(image_pixels, w_pattern1)}')
print(f'النمط 2: {simple_predict(image_pixels, w_pattern2)}')
```

### 4. دالة التنشيط ReLU (Non-linearity)
توضيح بسيط لكيفية قيام الشبكة بإهمال الإشارات الضعيفة أو السالبة (أقل من صفر).

```python
def relu(x):
    return np.maximum(0, x)

signals = np.array([-1.5, 0.4, -0.2, 2.1])
activated_signals = relu(signals)

print(f'الإشارات قبل التنشيط: {signals}')
print(f'الإشارات بعد ReLU (تم حذف القيم السالبة): {activated_signals}')
```

### 2.4 — الهندسة خلف العصبون: العصبون كخط فاصل

قبل الانتقال للبوابات المنطقية، يجب أن ندرك سراً رياضياً بسيطاً: **معادلة العصبون هي نفسها معادلة المستقيم.**

في الرياضيات، يُكتب المستقيم كـ $ax + by + c = 0$. وفي العصبون نكتب $w_1x_1 + w_2x_2 + b = 0$.

- الأوزان ($w$) تحدد ميل المستقيم واتجاهه.
- الانحياز ($b$) يحدد موقع المستقيم (إزاحته).

**ماذا يعني هذا؟**
يعني أن العصبون الواحد يعمل كـ "مقص"؛ فهو يرسم خطاً مستقيماً في البيانات، ويقول: "كل ما فوق هذا الخط ينتمي للفئة (أ)، وكل ما تحته ينتمي للفئة (ب)". هذا ما نسميه **الفصل الخطي (Linear Separation)**.

</div>

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. تحديد أوزان العصبون والانحياز (الوزن يحدد ميل الخط والانحياز يحركه)
w1, w2, b = 0.8, -0.5, -0.2

# 2. توليد نقاط على المحور الأفقي لحساب الخط
x_vals = np.linspace(-1, 2, 100)

# 3. معادلة المستقيم الخاص بالعصبون (مشتقة من: w1*x + w2*y + b = 0)
y_vals = -(w1 * x_vals + b) / w2

# 4. رسم الخط والمناطق
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, '--r', label='Neuron Decision Line')

# تلوين المنطقة فوق وتحت الخط
plt.fill_between(x_vals, y_vals, 2, alpha=0.2, color='blue', label='Class 1 (Active)')
plt.fill_between(x_vals, y_vals, -1, alpha=0.2, color='orange', label='Class 0 (Inactive)')

# 5. تحسين المظهر وإظهار الرسم
plt.title('How a Neuron Sees the World: Linear Separation')
plt.xlabel('Input 1 (x1)')
plt.ylabel('Input 2 (x2)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.show()
```

### 2.5 — التعلم وحل المنطق: بوابات (AND/OR) مقابل معضلة (XOR)

تعتبر البوابات المنطقية أفضل مثال لفهم قدرة الشبكات العصبية:

1. **بوابات AND و OR:** هي بوابات "خطية". انظر إلى توزيع النقاط؛ يمكنك رسم خط مستقيم واحد يفصل بين الصفر والواحد. لذا، يستطيع **عصبون واحد** تعلمها بسهولة عبر تعديل أوزانه وانحيازه.

2. **بوابة XOR (أو الحصرية):** هنا تكمن المعضلة التاريخية. إذا حاولت رسم خط واحد للفصل بين النتائج، ستجد الأمر مستحيلاً. تسمى هذه المشكلة **"عدم القابلية للفصل الخطي" (Non-linear Separability)**.

![Logic Gates Linear Separation](https://miro.medium.com/v2/resize:fit:1400/1*arBn7oCsLWbo71yS17nKRg.png)

### كيف يتم الحل؟
لحل XOR، نحتاج إلى "ثني" المساحة. يتم ذلك بإضافة **طبقة مخفية (Hidden Layer)**. العصبونات في الطبقة المخفية تقوم بتحويل المدخلات إلى تمثيل جديد يصبح فيه الفصل الخطى ممكناً في طبقة الخرج.

</div>

```python
# دالة التنشيط البسيطة (الخطوة) - مشتركة في الكودين
def step_function(x):
    return 1 if x >= 0 else 0

# العصبون المفرد
def single_neuron(x1, x2, w1, w2, bias):
    linear_sum = (x1 * w1) + (x2 * w2) + bias
    return step_function(linear_sum)

# الاحتمالات الأربعة للمدخلات
inputs = [[0, 0], [0, 1], [1, 0], [1, 1]]

# 1. اختبار بوابة AND
print("--- AND Gate Results ---")
for x in inputs:
    # الأوزان 1 و 1، والانحياز -1.5 (يجب أن يكون كلا المدخلين 1 ليتخطى الصفر)
    result = single_neuron(x[0], x[1], w1=1, w2=1, bias=-1.5)
    print(f"Input: {x} -> Output: {result}")

# 2. اختبار بوابة OR
print("\n--- OR Gate Results ---")
for x in inputs:
    # الأوزان 1 و 1، والانحياز -0.5 (أي مدخل قيمته 1 يكفي لتخطى الصفر)
    result = single_neuron(x[0], x[1], w1=1, w2=1, bias=-0.5)
    print(f"Input: {x} -> Output: {result}")
```

### 2.6 — حل مشكلة XOR باستخدام NumPy (شبكة من طبقتين)

كما ذكرنا، عصبون واحد لا يكفي. سنقوم الآن بمحاكاة شبكة صغيرة تحتوي على عصبونين في الطبقة المخفية وعصبون واحد في الخرج.

**الفكرة:** الطبقة المخفية ستقوم بعمل بوابتي (OR) و (NAND) معاً، ثم تقوم طبقة الخرج بدمج النتائج لتعطي (XOR).
![شبكة من طبقتين](https://i.ibb.co/PGcP4M74/Gemini-Generated-Image-oixsi4oixsi4oixs.jpg)
</div>

```python
# نفس دالة التنشيط البسيطة تماماً لضمان الموائمة
def step_function(x):
    return 1 if x >= 0 else 0

# دالة العصبون المفرد التي يفهمها الطلاب من الدرس السابق
def neuron(x1, x2, w1, w2, bias):
    linear_sum = (x1 * w1) + (x2 * w2) + bias
    return step_function(linear_sum)

# شبكة XOR المكونة من طبقتين (تستخدم عصبونات بسيطة)
def xor_network(x1, x2):
    # الطبقة المخفية:
    # العصبون الأول يمثل بوابة OR
    h1 = neuron(x1, x2, w1=1, w2=1, bias=-0.5)

    # العصبون الثاني يمثل بوابة NAND (عكس AND تماماً عبر عكس الإشارات)
    h2 = neuron(x1, x2, w1=-1, w2=-1, bias=1.5)

    # طبقة المخرج:
    # عصبون يدمج مخرجات الطبقة المخفية (يعمل كبوابة AND)
    output = neuron(h1, h2, w1=1, w2=1, bias=-1.5)

    return output

# المدخلات الأربعة للاختبار
inputs = [[0, 0], [0, 1], [1, 0], [1, 1]]

print("--- Resolved XOR Gate Results ---")
for x in inputs:
    result = xor_network(x[0], x[1])
    print(f"Input: {x} -> Output: {result}")
```

### 2.7 — تطبيق التمرير الخلفي (Backpropagation) والتعلم

الآن، بدلاً من وضع الأوزان يدوياً، سنقوم بتدريب الشبكة لتعلمها. سنتبع هذه الخطوات في كل دورة (Epoch):
1. **التمرير الأمامي:** حساب التوقعات.
2. **حساب الخطأ:** الفرق بين التوقع والنتيجة الحقيقية.
3. **التمرير الخلفي:** حساب مدى تأثير كل وزن في هذا الخطأ.
4. **تحديث الأوزان:** تعديل الأوزان باستخدام "معدل التعلم" (Learning Rate).

</div>

```python
import numpy as np

# 1. دالة التنشيط ومشتقتها
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(out):
    return out * (1 - out)

# بيانات XOR
inputs = [[0,0], [0,1], [1,0], [1,1]]
targets = [0, 1, 1, 0]

# 2. تهيئة أوزان مدروسة عشوائياً لضمان كسر التناظر والوصول للحل بسرعة
np.random.seed(10) # تثبيت العشوائية لضمان نفس النتيجة الممتازة في كل تشغيل

w11, w21, b1 = np.random.randn(), np.random.randn(), np.random.randn()
w12, w22, b2 = np.random.randn(), np.random.randn(), np.random.randn()
w_h1, w_h2, b_out = np.random.randn(), np.random.randn(), np.random.randn()

learning_rate = 0.3
epochs = 4000  # زيادة الدورات لضمان دقة التقارب التام

# 3. حلقة التدريب
for epoch in range(epochs):
    for i in range(4):
        x1, x2 = inputs[i][0], inputs[i][1]
        target = targets[i]

        # --- أولاً: التمرير الأمامي ---
        h1_out = sigmoid((x1 * w11) + (x2 * w21) + b1)
        h2_out = sigmoid((x1 * w12) + (x2 * w22) + b2)
        final_out = sigmoid((h1_out * w_h1) + (h2_out * w_h2) + b_out)

        # --- ثانياً: التمرير الخلفي ---
        error_out = target - final_out
        delta_out = error_out * sigmoid_derivative(final_out)

        delta_h1 = delta_out * w_h1 * sigmoid_derivative(h1_out)
        delta_h2 = delta_out * w_h2 * sigmoid_derivative(h2_out)

        # --- ثالثاً: تحديث الأوزان ---
        w_h1 += learning_rate * delta_out * h1_out
        w_h2 += learning_rate * delta_out * h2_out
        b_out += learning_rate * delta_out

        w11 += learning_rate * delta_h1 * x1
        w21 += learning_rate * delta_h1 * x2
        b1  += learning_rate * delta_h1

        w12 += learning_rate * delta_h2 * x1
        w22 += learning_rate * delta_h2 * x2
        b2  += learning_rate * delta_h2

# 4. طباعة النتائج النهائية بعد التعلم الكامل
print("--- Training Completed Successfully ---")
for i in range(4):
    x1, x2 = inputs[i][0], inputs[i][1]
    h1 = sigmoid((x1 * w11) + (x2 * w21) + b1)
    h2 = sigmoid((x1 * w12) + (x2 * w22) + b2)
    pred = sigmoid((h1 * w_h1) + (h2 * w_h2) + b_out)
    binary_class = 1 if pred > 0.5 else 0
    print(f"Input: [{x1}, {x2}] -> Raw Output: {pred:.4f} -> Class: {binary_class}")
```

3. **تحديد الأوزان يدويًا:** اختر قيمة مناسبة للوزن وقيمة للانحياز تحقق منطق بوابة NOT (تذكر أنك بحاجة إلى وزن يعكس التأثير).
4. **اختبار النموذج:** مرر القيم `0` و `1` داخل حلقة تكرارية (Loop) واطبع المدخل والمخرج على الشاشة للتأكد من صحة الحل.
